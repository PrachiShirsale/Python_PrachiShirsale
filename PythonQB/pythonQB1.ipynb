### 18.	Segregate file in small , medium and large size in different folder
### o	filesize less than and equal to 10KB is small file
### o	filesize more than 10 KB is the medium file
### o	filesize more than 40 kb is the large file
### o	create the log file 


```python
import shutil
import os
import logging

def createDir(directoryName):
    """
    Creates a directory if it does not exist.
    """
    try:
        if not os.path.exists(directoryName):
            os.mkdir(directoryName)
            logging.info(f"Directory created: {directoryName}")
    except Exception :
        logging.error(f"Error creating directory")

logging.basicConfig(
    filename="fileSorting.log",
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s"
)

baseDir = os.getcwd()
smallFiles = os.path.join(baseDir, "SmallFiles")
mediumFiles = os.path.join(baseDir, "MediumFiles")
largeFiles = os.path.join(baseDir, "LargeFiles")

createDir(smallFiles)
createDir(mediumFiles)
createDir(largeFiles)

def fileSorting(srcFolder):
    """
    Sorts files based on their size and moves them to designated folders.
    """
    try:
        if not os.path.exists(srcFolder):
            logging.error(f"Source folder does not exist: {srcFolder}")
            return
        
        for file in os.listdir(srcFolder):
            filePath = os.path.join(srcFolder, file)
            
            if os.path.isdir(filePath):
                logging.info(f"Skipping directory: {file}")
                continue
            
            sizeKB = os.path.getsize(filePath) // 1024
            logging.info(f"Processing file: {file}, Size: {sizeKB} KB")
            
            try:
                if sizeKB <= 10:
                    shutil.move(filePath, os.path.join(smallFiles, file))
                    logging.info(f"Moved {file} to SmallFiles")
                elif 10 < sizeKB <= 40:
                    shutil.move(filePath, os.path.join(mediumFiles, file))
                    logging.info(f"Moved {file} to MediumFiles")
                else:
                    shutil.move(filePath, os.path.join(largeFiles, file))
                    logging.info(f"Moved {file} to LargeFiles")
            except Exception :
                logging.error(f"Error moving file ")
    
    except Exception:
        logging.error(f"Error sorting files")

sourceFolder = "C:\\Users\\Admin\\Desktop\\PythonBizmetric"
fileSorting(sourceFolder)
print("File sorting completed. Check 'fileSorting.log' for details.")
```

### 20.	Ask user to enter the valid file path and check which word is having more repetition in the file

```python
import os

def getFilePath():
    """Prompt the user to enter a valid file path and check if it exists."""
    while True:
        filePath = input("Enter the file path: ").strip()
        if os.path.exists(filePath):
            return filePath
        else:
            print("Invalid file path")

def countWordFrequency(filePath):
    """Read the file and count the frequency of each word, ignoring case sensitivity."""
    wordCount = {}

    with open(filePath, 'r', encoding='utf-8') as file:
        for line in file:
            words = line.strip().split()
            for word in words:
                word = word.lower().strip(",.!?;:\"()[]{}")  
                if word:
                    wordCount[word] = wordCount.get(word, 0) + 1
    
    return wordCount

def findMostFrequentWord(wordCount):
    """Find the word with the highest frequency."""
    if not wordCount:
        return None, 0
    return max(wordCount.items(), key=lambda item: item[1])

filePath = getFilePath()
wordCount = countWordFrequency(filePath)
mostFrequentWord, maxCount = findMostFrequentWord(wordCount)

if mostFrequentWord:
    print(f" most repeated word is : {mostFrequentWord} : {maxCount} ")
else:
    print("No words found in the file")
```

### 24.	Ask user to enter the folder path. Folder may contain more than one file and with different extensions. Read only text file and return the more repetitive word

```python
import os
import sys

def getFldr():
    """Prompt user for a valid folder path and check existence."""
    while True:
        try:
            fldr = input("Enter the folder path: ").strip()
            if os.path.exists(fldr) and os.path.isdir(fldr):
                return fldr
            else:
                print("Invalid folder path")
        except:
            print(sys.exc_info()[1])

def getTxts(fldr):
    """Return a list of .txt files in the given folder."""
    try:
        return [os.path.join(fldr, f) for f in os.listdir(fldr) if f.endswith(".txt")]
    except:
        print("Error:", sys.exc_info()[1])
        return []

def countFreq(txts):
    """Read all text files and count word frequency."""
    wCnt = {}
    try:
        for file in txts:
            with open(file, 'r') as f:
                for line in f:
                    words = line.strip().split()
                    for w in words:
                        w = w.lower().strip(",.!?;:\"()[]{}")
                        if w:
                            wCnt[w] = wCnt.get(w, 0) + 1
    except:
        print("Error:", sys.exc_info()[1])
    return wCnt

def findMax(wCnt):
    """Find the most frequently occurring word."""
    try:
        return max(wCnt.items(), key=lambda x: x[1]) if wCnt else (None)
    except:
        print(sys.exc_info()[1])
        return None

fldr = getFldr()
txts = getTxts(fldr)

if txts:
    wCnt = countFreq(txts)
    maxWrd, maxCnt = findMax(wCnt)
    if maxWrd:
        print(f"Most repeated word: '{maxWrd}' : {maxCnt} ")
    else:
        print("No words found in files")
else:
    print("No text files found in the folder")
```

```python
import sys

def exceptionHandler(func):
    """Decorator to handle exceptions and print error messages."""
    def wrapper(*args, **kwargs):
        try:
            return func(*args, **kwargs)
        except Exception:
            print(sys.exc_info()[1]) 
    return wrapper

@exceptionHandler
def valName(name):
    """Function to validate a name. Raises error if invalid."""
    if not name.isalpha():
        raise ValueError("Invalid name Name should contain only letters.")
    return f"Valid Name: {name}"

print(valName("Prachi"))  
```

### 29.	Create a class pharma. One user A will enter the medicine name and price.  Another user B will generate the bill based on entry done by the first user

```python
import sys
import logging


logging.basicConfig(filename="pharma.log", level=logging.INFO,
                    format="%(asctime)s - %(levelname)s - %(message)s")

class Pharma:
    """Class to manage medicines and generate bills."""
    def __init__(self):
        """Initialize an empty medicine inventory."""
        self.medicineData = {}

    def addMedicine(self):
        """User A: Add medicines with prices via user input."""
        try:
            while True:
                name = input("Enter medicine name or type 'exit' to exit ").strip().lower()
                if name == "exit":
                    break
                if name in self.medicineData:
                    print(f"Medicine '{name}' already exists! Try another one.")
                    logging.warning(f"Cannot re-add medicine: {name}")
                    continue
                price = float(input(f"Enter price for '{name}': ").strip())
                if price <= 0:
                    raise ValueError("Price must be greater than zero.")
                self.medicineData[name] = price
                print(f"Medicine '{name}' added successfully!")
                logging.info(f"Added medicine: {name} | Price: ₹{price}")

        except ValueError:
            error_msg = str(sys.exc_info()[1])
            print(error_msg)
            logging.error(f"ValueError while adding medicine: {error_msg}")

    def generateBill(self):
        """User B: Generate bill based on selected medicines."""
        try:
            if not self.medicineData:
                print("No medicines, Please add medicines first.")
                logging.warning("Billing with no medicines available.")
                return
            selectedMeds = input("Enter medicines for billing ---> COMMA-SEPERATED : ").strip().lower().split(",")
            selectedMeds = [med.strip() for med in selectedMeds if med.strip()]

            total = 0
            print("--- Pharmacy Bill ---")
            for med in selectedMeds:
                if med in self.medicineData:
                    price = self.medicineData[med]
                    total += price
                    print(f"{med.capitalize()}: ₹{price}")
                else:
                    print(f"{med.capitalize()}: Not available")
                    logging.warning(f"Medicine not found in billing: {med}")

            gst = round(total * 0.10, 2)
            grandTotal = round(total + gst, 2)

            print(f"Total: ₹{total}")
            print(f"GST (10%): ₹{gst}")
            print(f"Grand Total: ₹{grandTotal}")
            logging.info(f"Bill Generated | Total: ₹{total} | GST: ₹{gst} | Grand Total: ₹{grandTotal}")
        except Exception:
            error_msg = str(sys.exc_info()[1])
            print(error_msg)
            logging.error(f"Error during billing: {error_msg}")

pharmacy = Pharma()
print("Adding Medicines... ")
pharmacy.addMedicine()
pharmacy.generateBill()


```

### 30.	Read the file if the file contains number, then return the dictionary based on position and values of the number.

```python
def extNums(fPath):
    numDict = {}
    try:
        with open(fPath, 'r') as f:
            wrds = f.read().split()
        pos = 1
        for w in wrds:
            try:
                num = float(w) if "." in w else int(w)
                numDict[num] = pos
            except ValueError:
                pass
            pos += 1

    except FileNotFoundError:
        print("Invalid file path.")
    except Exception:
        print(sys.exc_info()[1])
    return numDict

fPath = input("Enter file path: ").strip()
print("Extracted Numbers:", extNums(fPath))
```

### 31.	Create a class for file reading file if the file contains number, then return the dictionary based on user object, having sub dictionary position and values of the number against that user file.

```python
import sys
import logging

logger = logging.getLogger("FileProcessorLogger")
logger.setLevel(logging.INFO)
fHandler = logging.FileHandler("file_processor.log")
formatter = logging.Formatter("%(asctime)s - %(levelname)s - %(message)s")
fHandler.setFormatter(formatter)
logger.addHandler(fHandler)

class FileProcessor:
    def __init__(self, uName, fPath):
        self.uName = uName
        self.fPath = fPath
        self.numData = {}

    def extractNums(self):
        try:
            with open(self.fPath, 'r') as f:
                wrds = f.read().split()
            pos = 1
            for w in wrds:
                try:
                    num = float(w) if "." in w else int(w)
                    self.numData[num] = pos
                except ValueError:
                    pass
                pos += 1
            logger.info(f"File processed successfully for user: {self.uName}")

        except FileNotFoundError:
            err = "Invalid file path..."
            logger.error(err)
            print(err)
        except Exception:
            err = str(sys.exc_info()[1])
            logger.error(err)
            print(err)
        return {self.uName: self.numData}

uName = input("Enter user name : ").strip()
fPath = input("Enter file path : ").strip()

usrObj = FileProcessor(uName, fPath)
print("Extracted Data : ", usrObj.extractNums())
```

### 33.	Create a super class user. Having methods register, sing in, sign out.    Super class can be inherited by student, professor, admin
### Admin and students can view the marks and professor can enter the marks
### But marks can be visible to students on the given result date


```python
import logging
import datetime
import sys

logger = logging.getLogger("UserLogger")
logger.setLevel(logging.INFO)
fHandler = logging.FileHandler("user_activity.log")
formatter = logging.Formatter("%(asctime)s - %(levelname)s - %(message)s")
fHandler.setFormatter(formatter)
logger.addHandler(fHandler)

class User:
    users = {} 

    def __init__(self, name, role):
        self.name = name
        self.role = role
        self.loggedIn = False

    def register(self):
        try:
            if self.name in User.users:
                raise ValueError("User already registered")
            password = input(f"Set password for {self.name}: ").strip()
            if len(password) < 4:
                raise ValueError("Password must be at least 4 characters.")
            User.users[self.name] = password
            logger.info(f"{self.role} {self.name} registered successfully.")
            print("Registration successful!")
        except ValueError:
            logger.error("Value Error Occurred")
            print("Value Error Occurred")

    def signIn(self):
        try:
            password = input(f"Enter password for {self.name}: ").strip()
            if User.users.get(self.name) == password:
                self.loggedIn = True
                logger.info(f"{self.role} {self.name} signed in.")
                print(f"{self.name}, you are now signed in!")
            else:
                raise ValueError("Invalid credentials!")
        except ValueError:
            logger.error("Vlaue error occurred")
            print("Value Error occurred")

    def signOut(self):
        if self.loggedIn:
            self.loggedIn = False
            logger.info(f"{self.role} {self.name} signed out.")
            print(f"{self.name}, you are now signed out.")
        else:
            print("You are not logged in!")

class Professor(User):
    marksData = {} 

    def __init__(self, name):
        super().__init__(name, "Professor")

    def enterMarks(self):
        try:
            if not self.loggedIn:
                raise PermissionError("Sign in required to enter marks.")
            student = input("Enter Student Name: ").strip()
            marks = int(input(f"Enter marks for {student}: "))
            if not (0 <= marks <= 100):
                raise ValueError("Marks should be between 0 and 100.")

            resultDate = input("Enter result date (YYYY-MM-DD): ").strip()
            resultDate = datetime.datetime.strptime(resultDate, "%Y-%m-%d").date()

            Professor.marksData[student] = (marks, resultDate)
            logger.info(f"Marks entered: {student} - {marks} (Result Date: {resultDate})")
            print(f"Marks for {student} recorded successfully.")
        except ValueError :
            logger.error("Value Error Occurred")
            print("Value Error Occurred")
        except PermissionError as e:
            logger.error(e)
            print(e)

class Student(User):
    def __init__(self, name):
        super().__init__(name, "Student")

    def viewMarks(self):
        try:
            if not self.loggedIn:
                raise PermissionError("Sign in required to view marks.")
            if self.name not in Professor.marksData:
                raise ValueError("Marks not found for your name.")
            marks, resultDate = Professor.marksData[self.name]
            today = datetime.date.today()

            if today >= resultDate:
                logger.info(f"Student {self.name} viewed marks: {marks}")
                print(f"Your Marks: {marks}")
            else:
                daysLeft = (resultDate - today).days
                raise ValueError(f"Marks will be available in {daysLeft} days.")
        except ValueError as e:
            logger.error(e)
            print(e)
        except PermissionError as e:
            logger.error(e)
            print(e)

class Admin(User):
    def __init__(self, name):
        super().__init__(name, "Admin")

    def viewAllMarks(self):
        try:
            if not self.loggedIn:
                raise PermissionError("Sign in required to view all marks.")
            if not Professor.marksData:
                raise ValueError("No marks data available.")

            logger.info("Admin viewed all marks.")
            print("All Student Marks ")
            for student, (marks, resultDate) in Professor.marksData.items():
                print(f"{student}: {marks} (Result Date: {resultDate})")
        except ValueError:
            logger.error("Value Error Occurrred")
            print("Value error Occurred")
        except PermissionError as e:
            logger.error(e)
            print(e)

prof = Professor("Dr. Smith")
stud = Student("Alice")
admin = Admin("Mr. Admin")

prof.register()
stud.register()
admin.register()

prof.signIn()
prof.enterMarks()

stud.signIn()
stud.viewMarks()

admin.signIn()
admin.viewAllMarks()
```

### 34.	Create suitable classes, methods and attributes for resume exchange. Employee to upload the file and the same file is visible to employer 

```python
import logging
import uuid

logger = logging.getLogger("ResumeExchangeLogger")
logger.setLevel(logging.INFO)
fHandler = logging.FileHandler("resume_exchange.log")
formatter = logging.Formatter("%(asctime)s - %(levelname)s - %(message)s")
fHandler.setFormatter(formatter)
logger.addHandler(fHandler)

class Employee:
    employee_data = {}  
    def __init__(self, name, email):
        self.name = name
        self.email = email
        self.emp_id = str(uuid.uuid4())[:8]  
        self.resume_link = None
        Employee.employee_data[self.emp_id] = {
            "name": self.name,
            "email": self.email,
            "resume_link": None
        }
        logger.info(f"New Employee registered: {self.name}, ID: {self.emp_id}")

    def upload_resume(self, resume_link):
        try:
            if not resume_link.startswith("http"):
                raise ValueError("Invalid Resume Link! Must start with 'http'.")
            Employee.employee_data[self.emp_id]["resume_link"] = resume_link
            self.resume_link = resume_link
            logger.info(f"Resume uploaded for Employee ID: {self.emp_id}")
            print("Resume uploaded successfully!")
        except ValueError as e:
            logger.error(e)
            print(e)

    def get_employee_id(self):
        return self.emp_id

class Employer:
    @staticmethod
    def view_resume(emp_id):
        try:
            if emp_id not in Employee.employee_data:
                raise KeyError("Invalid Employee ID! No such employee found.")
            employee = Employee.employee_data[emp_id]
            if not employee["resume_link"]:
                raise ValueError(f"No resume uploaded for Employee ID: {emp_id}.")

            logger.info(f"Employer accessed resume of Employee ID: {emp_id}")
            print(f"Employee Name: {employee['name']}\nResume Link: {employee['resume_link']}")
        except KeyError:
            logger.error("Key Error")
            print("key Error ")
        except ValueError :
            logger.error("Value Error ")
            print("Value Error")

emp1 = Employee("Prachi Shirsale", "prachi@gmail.com")
emp2 = Employee("Rahul Smith", "rahul@gmail.com")

print(f"Prachi's Employee ID: {emp1.get_employee_id()}")
print(f"Rahuls's Employee ID: {emp2.get_employee_id()}")

emp1.upload_resume("http://example.com/prachi_resume.pdf")
emp2.upload_resume("http://example.com/rahul_resume.pdf")

Employer.view_resume(emp1.get_employee_id())  
Employer.view_resume(emp2.get_employee_id())  
```

```python
# SQL 
CREATE TABLE employees (
    emp_id INT IDENTITY(1,1) PRIMARY KEY,
    name NVARCHAR(255) NOT NULL,
    mobile NVARCHAR(15) UNIQUE NOT NULL,
    address NVARCHAR(MAX) NOT NULL,
    salary DECIMAL(10,2) NOT NULL
);

CREATE TABLE leave_requests (
    leave_id INT IDENTITY(1,1) PRIMARY KEY,
    emp_id INT NOT NULL,
    start_date DATE NOT NULL,
    end_date DATE NOT NULL,
    reason NVARCHAR(MAX) NOT NULL,
    status NVARCHAR(20) DEFAULT 'Pending',
    FOREIGN KEY (emp_id) REFERENCES employees(emp_id) ON DELETE CASCADE
);

CREATE TABLE approved_leaves (
    leave_id INT IDENTITY(1,1) PRIMARY KEY,
    emp_id INT NOT NULL,
    start_date DATE NOT NULL,
    end_date DATE NOT NULL,
    reason NVARCHAR(MAX) NOT NULL,
    FOREIGN KEY (emp_id) REFERENCES employees(emp_id) ON DELETE CASCADE
);

```
