```python
import shutil
import os

def createDir(dirName):
    ''' 
    Creates directory if directory Dosen't Exists
    '''
    if not os.path.exists(dirName):
        os.mkdir(dirName)

baseDir = os.getcwd()
smallFiles = os.path.join(baseDir, "Small Files")
mediumFiles = os.path.join(baseDir, "Medium Files")
largeFiles = os.path.join(baseDir, "Large Files")

createDir(smallFiles)
createDir(mediumFiles)
createDir(largeFiles)

srcFolder = "C:\\Users\\Admin\\Desktop\\Python_Bizmetric"  

def FileSorting(srcFolder):
    '''
    Sort files based on their size and move them to specific folders
    '''
    for file in os.listdir(srcFolder):
        filePath = os.path.join(srcFolder, file) 
        try:
            if os.path.isdir(filePath):
                print(f"Skipping directory: {file}")
                continue  

            size = os.path.getsize(filePath) // 1024 
 
            print(f"Processing file: {file}, Size: {size} KB")

            if 5 <= size <= 100:
                shutil.move(filePath, os.path.join(smallFiles, file))
                print(f"Moved {file} to Small Files")
            elif 101 <= size <= 500:
                shutil.move(filePath, os.path.join(mediumFiles, file))
                print(f"Moved {file} to Medium Files")
            elif size >= 501:
                shutil.move(filePath, os.path.join(largeFiles, file))
                print(f"Moved {file} to Large Files")
            else:
                print(f"Skipping {file} : Size is below 5 KB")
        except Exception as e:
            print(f"Error occurred while processing file: {file}")
            print(e)

if os.path.exists(srcFolder):
    FileSorting(srcFolder)
    print("Files sorted successfully.")
else:
    print(f"Source folder '{srcFolder}' does not exist")
```
