```python
# DAY 2 Taks 1 - Generate a Hotel bill 

from functools import reduce

def generate_bill():
    ''' Takes Input as menu items and its quantity and generats the Total Bill according to its price '''
    try:
        Hotel_name = "Swastik"
        print('-' * 80)
        print(f"{Hotel_name:^80}")
        print(" ")
        print('-' * 80)

        sr_no = 'Sr.no'
        item = 'Item'
        ppp = 'Price_Per_Item'
        qnt = 'Quantity'
        TP = 'Total price'

        Item_Price = {'Iidli': 50, 'Pav Bhaji': 100, "Maggie": 50, "Dosa": 50}

        print(f"{sr_no:>10} {item:>10} {ppp:>20} {qnt:>15} {TP:>20}")
        print(" ")
        Sr_no = 1
        TotalPrice = []

        while True:
            Item = input("Enter Item To Order (or type 'end' to finish): ").strip()
            if Item.lower() == 'end':
                break

            if Item not in Item_Price:
                print("Invalid item. Please enter a valid menu item.")
                continue

            PPP = Item_Price[Item]
            try:
                InputQuantity = int(input("Enter Quantity: "))
                if InputQuantity <= 0:
                    print("Quantity must be a positive integer. Please try again.")
                    continue
            except ValueError:
                print("Invalid input. Quantity must be an integer. Please try again.")
                continue

            TotalP = PPP * InputQuantity
            TotalPrice.append(TotalP)
            print(f"{Sr_no:>10} {Item:>10} {PPP:>20} {InputQuantity:>15} {TotalP:>20}")
            print(" ")
            Sr_no += 1

        if TotalPrice:
            FinalBill = reduce(lambda x, y: x + y, TotalPrice)
        else:
            FinalBill = 0

        FB = "Total Bill"
        print('-' * 80)
        print(f"{FB:>60}{FinalBill:>18}")
    except Exception as e:
        print(f"An error occurred: {e}")

generate_bill()
```
